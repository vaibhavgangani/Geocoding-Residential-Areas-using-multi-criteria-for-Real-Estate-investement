```python
#Reading our previously filtered data CSV file
df = pd.read_csv(r"D:\GIS Python Assignment\landundervalue2.csv")
```

```python
df.head()
```

```python
#Creating a new column ADDRESS by concatenating 5 columns

df['ADDRESS'] = df['PAON'] + ',' + \
                df['Street'] + ',' + \
                df['Town/City'] + ',' + \
                df['District'] + ',' + \
                df['County'] 
```

```python
#Printing newly created information with ADDRESS Column

df.head()
```

```python
# Geocoding the ADDRESS column by using function and store it in a new POINT column

from geopy.extra.rate_limiter import RateLimiter
geocode = RateLimiter(locator.geocode, min_delay_seconds=1)
df['location'] = df['ADDRESS'].apply(geocode)
df['point'] = df['location'].apply(lambda loc: tuple(loc.point) if loc else None)
```

```python
#printing it
df.head()
```

```python
# check information at 1st index in point column
df['point'][0][0]
```

```python
# split point column into latitude, longitude and altitude columns
df[['latitude', 'longitude', 'altitude']] = pd.DataFrame(df['point'].tolist(), index=df.index)
df.head()
```

```python
#check the sum, if any latitude is Null

df.latitude.isnull().sum()
```

```python
df = df[pd.notnull(df["latitude"])]
```

```python
# using folium, plot the coordinate on a OSM basemap

map1 = folium.Map(
    location=[52.4862, -1.8904],
    tiles='OpenStreetMap',
    zoom_start=6,
)
```

```python
#styling the data
df.apply(lambda row:folium.CircleMarker(location=[row["latitude"], row["longitude"]]).add_to(map1), axis=1)
map1
```

```python
#Save the output as a seperate CSV file

map1.save(r"D:\GIS Python Assignment\SuitableLands2.html")
```
